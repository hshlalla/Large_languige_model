```python
from rapidfuzz import fuzz
import pandas as pd
import re

# 한글, 한자, 일본어 정규식 패턴
hangul_hanja_japanese_pattern = re.compile(r"[ㄱ-ㅎ가-힣一-龥ぁ-ゔァ-ヴー々〆〤]")

# 결과를 저장할 리스트
results = []

# 반복하면서 유사도 측정
for i in range(len(precessed_model_df)):
    label = precessed_model_df["modelID"][i]
    std_name = precessed_model_df["modelSTDName"][i]
    aliases = precessed_model_df["modelAlias"][i].split(",")

    for alias in aliases:
        score = fuzz.partial_ratio(std_name, alias)

        # 90점 이상은 저장하지 않음
        if 20 < score < 40:
            results.append({
                    "modelSTDName": std_name,
                    "alias": alias,
                    "score": score,
                })
        
        # 0점일 경우에만 한글, 한자, 일본어 포함 여부 확인 후 저장
        if score == 0:
            if hangul_hanja_japanese_pattern.search(alias):
                results.append({
                    "modelSTDName": std_name,
                    "alias": alias,
                    "score": score,
                })

# 결과를 DataFrame으로 변환
result_df = pd.DataFrame(results)
result_df
```
